```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 7, 8]
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)


plt.xlabel('EJE X')
plt.ylabel('EJE Y')


plt.show()
```

```python
# grafico de dispersion
plt.scatter(x,y)

plt.show()
```

```python
#histogramas
import numpy as np


datos =np.random.randn(1000)

plt.hist(datos, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('valor')
plt.ylabel('frecuencia')
plt.show()
```

```python
#diagrama de barras

categorias =['A','B','C','D']
valor =[30, 40, 50, 38]

plt.bar(categorias, valor, color='darkgreen')

plt.show()
```
